### Handling Missing Values

Before proceeding with the analysis, it's crucial to address any missing values in the dataset. Missing values can distort our analysis and lead to inaccurate conclusions. From the `britannia.info()` output, we identified that 'Trades', 'Deliverable Volume', and '%Deliverble' columns contain `NaN` values. The code above calculates and displays the count and percentage of missing values for each column.

Upon inspection, it's evident that 'Trades' has a significant portion of its data missing (approximately 53.69%), while 'Deliverable Volume' and '%Deliverble' have about 9.59% missing. Depending on the analysis goals, different strategies can be employed: dropping rows/columns, imputation (mean, median, mode), or more advanced methods.

### Missing Value Handling Strategy Implementation

Given the high percentage of missing values in the 'Trades' column (approximately 53.69%), dropping this column seems appropriate as imputation might introduce too much bias. For 'Deliverable Volume' and '%Deliverble', which have a smaller percentage of missing values (around 9.59%), median imputation is a reasonable strategy. The median is chosen over the mean to be less sensitive to outliers in the data.

```python
# Drop the 'Trades' column
britannia_cleaned = britannia.drop(columns=['Trades'])

# Impute missing values in 'Deliverable Volume' and '%Deliverble' with their medians
median_deliverable_volume = britannia_cleaned['Deliverable Volume'].median()
median_deliverble_percentage = britannia_cleaned['%Deliverble'].median()

britannia_cleaned['Deliverable Volume'] = britannia_cleaned['Deliverable Volume'].fillna(median_deliverable_volume)
britannia_cleaned['%Deliverble'] = britannia_cleaned['%Deliverble'].fillna(median_deliverble_percentage)

print('Missing values after handling:')
display(britannia_cleaned.isnull().sum())
```

## Asking and Answering Questions

In this section, we will formulate and answer specific questions about the BRITANNIA stock dataset. By leveraging our cleaned data and the powerful capabilities of `pandas` for data manipulation and `matplotlib`/`seaborn` for visualization, we aim to extract deeper insights into the stock's performance and trends. Each question will be addressed with relevant code and a clear explanation of the findings.

#### Q1: What is the total percentage change in BRITANNIA's closing price from its earliest recorded date to its latest?

```python
# Get the first and last closing prices
first_close_price = britannia_cleaned['Close'].iloc[0]
last_close_price = britannia_cleaned['Close'].iloc[-1]

# Calculate the percentage change
percentage_change = ((last_close_price - first_close_price) / first_close_price) * 100

print(f"BRITANNIA's closing price on {britannia_cleaned['Date'].iloc[0].strftime('%Y-%m-%d')}: {first_close_price:.2f}")
print(f"BRITANNIA's closing price on {britannia_cleaned['Date'].iloc[-1].strftime('%Y-%m-%d')}: {last_close_price:.2f}")
print(f"Total percentage change in closing price: {percentage_change:.2f}%")
```

#### Answer to Q1: Total Percentage Change in Closing Price

To answer this question, we first identified the `Close` price from the very first trading day `(iloc[0])` and the very last trading day `(iloc[-1])` in our `britannia_cleaned` DataFrame. We then used the standard percentage change formula: `((last_value - first_value) / first_value) * 100`.

The result indicates a substantial increase in BRITANNIA's stock price over the period covered by the dataset. This positive percentage change reflects significant growth in the stock's value from its inception in the dataset until the most recent trading day, aligning with the upward trend observed in the 'Closing Price Over Time' plot.

#### Q2: Which dates had the highest and lowest trading volumes for BRITANNIA stock?

```python
# Find the row with the highest trading volume
highest_volume_row = britannia_cleaned.loc[britannia_cleaned['Volume'].idxmax()]
highest_volume_date = highest_volume_row['Date'].strftime('%Y-%m-%d')
highest_volume = highest_volume_row['Volume']

# Find the row with the lowest trading volume
lowest_volume_row = britannia_cleaned.loc[britannia_cleaned['Volume'].idxmin()]
lowest_volume_date = lowest_volume_row['Date'].strftime('%Y-%m-%d')
lowest_volume = lowest_volume_row['Volume']

print(f"Date with highest trading volume: {highest_volume_date} with {highest_volume} shares traded.")
print(f"Date with lowest trading volume: {lowest_volume_date} with {lowest_volume} shares traded.")
```

#### Answer to Q2: Dates with Highest and Lowest Trading Volumes

To answer this, we used the `idxmax()` and `idxmin()` pandas DataFrame methods to find the index of the maximum and minimum values in the 'Volume' column, respectively. Then, `loc` was used to retrieve the entire row for these indices, allowing us to extract the corresponding 'Date' and 'Volume'. The `strftime('%Y-%m-%d')` method was used to format the datetime objects into a readable date string.

The results highlight specific days with extreme trading activity, which could correlate with significant company announcements, financial reports, or major market movements. Identifying these dates can be crucial for further historical analysis to understand the drivers behind such high or low investor interest.

#### Q3: What is the average daily return of BRITANNIA stock, and how are these returns distributed?

```python
# Calculate daily returns
britannia_cleaned['Daily_Return'] = britannia_cleaned['Close'].pct_change() * 100

# Calculate the average daily return
average_daily_return = britannia_cleaned['Daily_Return'].mean()

print(f"Average Daily Return: {average_daily_return:.2f}%")

# Plot histogram of daily returns
plt.figure(figsize=(10, 6))
sns.histplot(britannia_cleaned['Daily_Return'].dropna(), kde=True, bins=50)
plt.title('Distribution of BRITANNIA Stock Daily Returns')
plt.xlabel('Daily Return (%)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

#### Answer to Q3: Average Daily Return and Distribution

To address this question, we first computed the 'Daily_Return' for the stock using the `.pct_change()` method on the 'Close' price, multiplying by 100 to express it as a percentage. This function calculates the percentage change between the current and a prior element. We then calculated the `mean()` of these daily returns to find the average.

The average daily return provides a general sense of the stock's daily performance. The histogram visualizes the frequency of different daily return percentages. A bell-shaped curve centered around the mean (or close to zero) is often expected for stock returns, but the tails of the distribution can indicate periods of high volatility (large positive or negative returns). This plot helps in understanding the typical range and likelihood of daily price movements.

#### Q4: How do the 50-day and 200-day Simple Moving Averages (SMA) compare to the closing price, and what trends do they indicate?

```python
# Calculate 50-day Simple Moving Average (SMA)
britannia_cleaned['SMA_50'] = britannia_cleaned['Close'].rolling(window=50).mean()

# Calculate 200-day Simple Moving Average (SMA)
britannia_cleaned['SMA_200'] = britannia_cleaned['Close'].rolling(window=200).mean()

# Plotting Close Price, 50-day SMA, and 200-day SMA
plt.figure(figsize=(14, 7))
plt.plot(britannia_cleaned['Date'], britannia_cleaned['Close'], label='Close Price', color='blue', alpha=0.7)
plt.plot(britannia_cleaned['Date'], britannia_cleaned['SMA_50'], label='50-Day SMA', color='red', alpha=0.7)
plt.plot(britannia_cleaned['Date'], britannia_cleaned['SMA_200'], label='200-Day SMA', color='green', alpha=0.7)

plt.title('BRITANNIA Stock: Close Price vs. 50-Day & 200-Day SMAs')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.grid(True)
plt.show()
```

#### Answer to Q4: 50-day and 200-day Simple Moving Averages

To answer this question, we computed two common technical indicators: the 50-day Simple Moving Average (SMA) and the 200-day Simple Moving Average (SMA). The `.rolling(window=n).mean()` method from pandas is used to calculate the mean of the `Close` price over a specified `n`-day window. SMAs are used to smooth out price data over time, helping to identify trends. A 50-day SMA is considered a short-term to medium-term trend indicator, while a 200-day SMA is a long-term trend indicator.

On the plot, we can observe the following:
*   When the 50-day SMA crosses above the 200-day SMA, it's often considered a 'golden cross', suggesting a potential bullish (upward) trend. We can see this occurring at several points, indicating periods of sustained growth.
*   Conversely, when the 50-day SMA crosses below the 200-day SMA, it's known as a 'death cross', suggesting a potential bearish (downward) trend. These crossovers are less frequent but can be seen during market downturns.
*   The `Close` price tends to oscillate around these moving averages. When the price is consistently above both SMAs, it signals a strong upward trend, and when it's below, it signals a downward trend.

These moving averages help in understanding the underlying trend by filtering out short-term fluctuations, making it easier to see the direction of the stock's movement over different time horizons.

#### Q5: What is the average monthly closing price of BRITANNIA stock, and are there any noticeable monthly patterns?

```python
# Extract year and month from the Date column
britannia_cleaned['Year'] = britannia_cleaned['Date'].dt.year
britannia_cleaned['Month'] = britannia_cleaned['Date'].dt.month

# Calculate the average monthly closing price
average_monthly_close = britannia_cleaned.groupby(['Year', 'Month'])['Close'].mean().reset_index()

# Plotting average monthly closing price
plt.figure(figsize=(16, 8))
sns.lineplot(data=average_monthly_close, x='Month', y='Close', hue='Year', palette='viridis')
plt.title('BRITANNIA Stock: Average Monthly Closing Price Over Years')
plt.xlabel('Month')
plt.ylabel('Average Close Price')
plt.xticks(ticks=range(1, 13), labels=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.legend(title='Year', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.show()
```

#### Answer to Q5: Average Monthly Closing Price and Patterns

To answer this question, we first extracted the `Year` and `Month` components from the `Date` column using the `.dt.year` and `.dt.month` accessors provided by pandas for datetime objects. Then, we used the `.groupby()` method to group the data by `Year` and `Month` and calculated the `mean()` of the `Close` price for each group. The `.reset_index()` method was used to convert the grouped output back into a DataFrame.

The line plot visualizes the average monthly closing price, with different lines representing different years. This allows us to observe if there are any consistent seasonal patterns in the stock's performance or if certain months typically show stronger or weaker performance.

From the plot, we can generally observe an upward trend in average monthly closing prices over the years, which aligns with the overall growth seen in the stock. While specific monthly patterns might not be strictly consistent across all years due to market dynamics, one might look for:
*   **Growth trends:** Are prices generally higher towards the end of the year than at the beginning?
*   **Seasonal dips/rises:** Do prices tend to dip in specific months (e.g., May or September, often cited as weaker months for stocks) or rise in others (e.g., year-end rallies)?

By comparing the lines for different years, we can see the stock's consistent upward trajectory, with each subsequent year generally starting and ending at a higher average monthly price than the previous one, albeit with fluctuations within each year.
